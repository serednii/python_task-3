```python
import random
import time
class Node:
    
    arrStrValue  = []
    arrStrChar = []
    
    @staticmethod
    def printTree(node):
        # node.widthTreeLeftRight()
        depthTree, lengthValues, l = node.widthTreeLeftRight()
        Node.arrStrValue = [' '* (lengthValues )] * depthTree
        Node.arrStrChar = [' '* (lengthValues )] * depthTree
        node.printTreeInList(0, True)        
        
        for s in range(len(Node.arrStrValue)):
            print(Node.arrStrValue[s])
            print(Node.arrStrChar[s])
    
    def __init__(self, value = None):
        self.value = value
        self.left = None # Посилання на лівий глибший обєкт
        self.right = None # Посилання на правий глибший обєкт
        self.widthLeft = 0 # Ширина дерева значень з ліва
        self.widthRight = 0 # Ширина дерева значень з права
        self.depth = 0 # Глубина дерева з гори в низ починаючи з 0
        
    
    def insert(self, value):
        if self.value is None:
            self.value = value
            return
        
        if value > self.value:
            if self.right is not None:
                self.right.insert(value)
            else:
                self.right = Node(value) 
                
                
        elif value < self.value:
            if self.left is not None:
                self.left.insert(value)
            else:
                self.left = Node(value)

    def widthTreeLeftRight(self):
        if self is None:
            return [0, 0, 0]
            
        l = 0
        r = 0
        deptchLeft = 0
        deptchRight = 0
        lengthValuesLeft = 0
        lengthValuesRight = 0
        lenNodeValue = len(str(self.value))
        
        if self.left is not None:
            self.left.depth =  self.depth + 1 #Прописуємо рівні строк
            deptchLeft, lengthValuesLeft, l = self.left.widthTreeLeftRight()
            
    
        if self.right is not None:
            self.right.depth =  self.depth + 1 #Прописуємо рівні строк
            deptchRight, lengthValuesRight, r = self.right.widthTreeLeftRight()
        
        
        self.widthLeft = l
        self.widthRight = r
        
        returnWidthpSubTree = l + r + lenNodeValue
        deptch = 1+ max(deptchLeft, deptchRight)
        lengthValues = lengthValuesLeft + lengthValuesRight + lenNodeValue # Загальна довжина значень всіх елементів з правої та лівої сторони та обєкта в якому ми знаходимося 
        return [deptch, lengthValues, returnWidthpSubTree]
        
        
        #****************************************************************************************
    
     #***************************************START WIDTHTREEE*************************************************************   widthTree
    def printTreeInList(self, widthParams = 0, flagStart = False):

        leftNewWidthParams = 0
        newleftParams = 0
        newRightParams = 0
        
        lenNodeValue = len(str(self.value))
        nodeLeftValue = len(str(self.left.value)) if self.left is not None else 0 # Тернарний оператор
        nodeRightValue = len(str(self.right.value)) if self.right is not None else 0
        nodeLeftWidthRight = self.left.widthRight if self.left is not None else 0
        nodeRightWidthLeft = self.right.widthLeft if self.right is not None else 0

        if flagStart:
            leftNewWidthParams =  self.widthLeft - nodeLeftValue // 2 - nodeLeftWidthRight
            newleftParams = self.widthLeft - nodeLeftValue  - nodeLeftWidthRight
            newRightParams = self.widthLeft + lenNodeValue + nodeRightWidthLeft
            # print('getRightParams start',  newleftParams, newRightParams, leftNewWidthParams, self.value)
        else:
            leftNewWidthParams = widthParams - nodeLeftValue // 2 - nodeLeftWidthRight
            newleftParams = widthParams - nodeLeftValue  - nodeLeftWidthRight
            newRightParams = widthParams + lenNodeValue + nodeRightWidthLeft
            # print('getRightParams next', newleftParams, newRightParams, leftNewWidthParams, self.value)
            
        
        
        #***************************PRINT STR**********************************
        number_left = nodeLeftValue//2-1 + nodeLeftWidthRight
        number_right = nodeRightWidthLeft + nodeRightValue//2-1
        strReplace = '_' * number_left  + str(self.value) + '_' * (number_right)
        
        Node.arrStrValue[self.depth] =  Node.arrStrValue[self.depth][:leftNewWidthParams]      + strReplace +       Node.arrStrValue[self.depth][leftNewWidthParams + len(strReplace):] 
        
        charSlechLeft = '/' if self.left is not None else ' ' # Тернарний оператор
        charSlechRight = '\\' if self.right is not None else ' '
        # Node.arrStrChar[self.depth-1] =  Node.arrStrChar[self.depth-1][:leftNewWidthParams]   + charSlechLeft +      Node.arrStrChar[self.depth-1][leftNewWidthParams + 1:leftNewWidthParams + len(strReplace)+1]   + charSlechRight +      Node.arrStrChar[self.depth-1][leftNewWidthParams + len(strReplace)+2:]
        leftNewWidthParams = leftNewWidthParams - 1 if leftNewWidthParams -1 >= 0 else 0

        Node.arrStrChar[self.depth] =  Node.arrStrChar[self.depth][:leftNewWidthParams]   + charSlechLeft +      ' ' * (len(strReplace))   + charSlechRight +      Node.arrStrChar[self.depth][leftNewWidthParams + (len(strReplace))+1:]
        #**********************************************************************
        
        
        if self.left is not None: 
            # print('getleftParams ', getleftParams, leftNewWidthParams, node.value)
            self.left.printTreeInList(newleftParams)
            
        if self.right is not None: 
            # print('getRightParams ', getRightParams, node.value)
            self.right.printTreeInList(newRightParams)
        return 
     #***************************************END WIDTHTREEE*************************************************************   
    





# lists = [12, 92, 79, 81, 20, 71, 37, 9, 73, 69, 91, 8, 77, 76, 78, 40, 65, 39, 10, 52, 64, 10, 18, 52, 6, 26, 28, 66, 16, 45]
# lists = [48, 86, 13, 56, 66, 35, 72, 11, 46, 54, 92, 77, 69, 37, 47, 92, 13, 15, 90, 7, 80, 5, 36, 90, 76, 90, 47, 90, 39, 21]
# lists = [17, 44, 70, 21, 73, 33, 57, 82, 9, 12, 54, 4, 38, 35, 19, 51, 48, 87, 56, 71, 91, 87, 99, 44, 50, 63, 77, 23, 5, 46]
# lists = [17, 44, 70, 21, 73, 33, 57, 82, 9,  54, 4, 38, 35, 19, 51, 48, 87, 91, 87, 99, 44, 23, 5, 46]
# lists = [17, 35, 20, 29, 38, 33, 10, 35, 5, 11, 21, 16, 13, 43, 17, 27, 9, 30, 46, 49]
# lists = [30, 36, 1, 15, 47, 9, 6, 14, 10, 50, 10, 49, 23, 13, 27, 19, 40, 6, 24, 45]
# lists = [17, 35, 20, 29, 38, 33, 10, 35, 5, 12]
# lists = [30, 36,  15, 47, 14, 10, 50, 10]
lists = [814818, 140754, 145119, 468835, 128813, 123114, 148665, 218626, 152211, 409582, 102558, 344502, 151117]


# d = Node()

# for n in lists:
#     print(n, end=", ")
#     d.insert(n)
# print()


for n in range(2):    
    d = Node()
    for i in range(50):
        d.insert((random.randint(0, 300)))
    Node.printTree(d)
       




    

```
